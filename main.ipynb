```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.cluster import KMeans, AffinityPropagation, MeanShift, AgglomerativeClustering, DBSCAN

from sklearn.preprocessing import StandardScaler
```

```python
file = "s1.txt"
data = pd.read_csv(file, delimiter=r"\s+", header=None)
X = data.values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
def plot_clusters(X, labels, title):
    plt.figure(figsize=(10, 12))
    sb.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='deep', legend=True).legend_.set_title("Clusters")
    plt.title(title)
    plt.show()
```

### KMeans

```python
kmeans = KMeans(n_clusters=15)
kmeans_labels = kmeans.fit_predict(X_scaled)
plot_clusters(X_scaled, kmeans_labels, 'KMeans')
```

### Affinity Propagation

```python
affinity = AffinityPropagation(damping=0.9, preference=-50)
affinity_labels = affinity.fit_predict(X_scaled)
plot_clusters(X_scaled, affinity_labels, "AffinityPropagation")
```

### Mean Shift

```python
mean_shift = MeanShift(bandwidth=0.3)
mean_shift_labels = mean_shift.fit_predict(X_scaled)
plot_clusters(X_scaled, mean_shift_labels, "MeanShift")
```

### Agglomerative Clustering

```python
agg_clustering = AgglomerativeClustering(n_clusters=15)
agg_labels = agg_clustering.fit_predict(X_scaled)
plot_clusters(X_scaled, agg_labels, "AgglomerativeClustering")
```

### DBSCAN

```python
dbscan = DBSCAN(eps=0.13, min_samples=15)
dbscan_labels = dbscan.fit_predict(X_scaled)
plot_clusters(X_scaled, dbscan_labels, "DBSCAN")
```
